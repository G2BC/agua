# Tempo de Processamento da Filogenia

```python
import os
import subprocess
import time
```

```python
# DiretÃ³rio principal onde vocÃª possui vÃ¡rias pastas
diretorio_principal = '/home/m_souza/tcc/dataset/00_new_dataset_vagner/5tipos/output'

# Iterar sobre as pastas no diretÃ³rio principal
for pasta in os.listdir(diretorio_principal):
    pasta_path = os.path.join(diretorio_principal, pasta)

    # Verificar se Ã© um diretÃ³rio
    if os.path.isdir(pasta_path):
        # Procurar pelo arquivo sequencias_treinamento.fasta na pasta
        arquivo_sequencias = os.path.join(pasta_path, 'sequencias_treinamento.fasta')

        # Verificar se o arquivo existe
        if os.path.exists(arquivo_sequencias):
            print(f'Analisando sequÃªncias em {pasta}')

            # Comando a ser executado
            comando = f"iqtree2 -s {arquivo_sequencias} -fast -m GTR+G4+F -alrt 1000 -nt AUTO"

            # InÃ­cio da contagem do tempo
            start_time = time.time()

            # Executar o comando usando subprocess
            process = subprocess.Popen(comando, shell=True)
            process.communicate()  # Aguardar a conclusÃ£o do processo

            # Fim da contagem do tempo
            end_time = time.time()

            # Calcular o tempo de execuÃ§Ã£o
            execution_time = end_time - start_time

            # Exibir o tempo de execuÃ§Ã£o
            print(f"Tempo de execuÃ§Ã£o em {pasta}: {execution_time} segundos")

            # Salvar as informaÃ§Ãµes em um arquivo tempo_de_execucao.txt
            arquivo_tempo_execucao = os.path.join(pasta_path, 'tempo_de_execucao.txt')
            with open(arquivo_tempo_execucao, "w") as file:
                file.write(f"Tempo de execuÃ§Ã£o: {execution_time} segundos")

print('ConcluÃ­do!')
```
