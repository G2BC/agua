# Tempo de Processamento da Filogenia

```python
import subprocess
import time

# Path das Sequências
path_sequencias = '/home/m_souza/tcc/dataset/dataset_agua/iqtree/iqtree_teste_tempo/teste/sequencias_treinamento.fasta'
# Comando a ser executado
comando = f"iqtree2 -s {path_sequencias} -fast -m GTR+G4+F -alrt 1000 -nt AUTO"

# Início da contagem do tempo
start_time = time.time()

# Executar o comando usando subprocess
process = subprocess.Popen(comando, shell=True)
process.communicate()  # Aguardar a conclusão do processo

# Fim da contagem do tempo
end_time = time.time()

# Calcular o tempo de execução
execution_time = end_time - start_time

# Exibir o tempo de execução
print(f"Tempo de execução: {execution_time} segundos")

# # Salvar as informações em um arquivo
# with open("tempo_de_execucao.txt", "w") as file:
#     file.write(f"Tempo de execução: {execution_time} segundos")
```
